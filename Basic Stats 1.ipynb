#### Insights


**Volume:**  
The distribution of volume is right-skewed, with most transactions involving a small number of units.

---

**Average Price:**  
The average price per unit varies significantly across transactions, with a noticeable right skew. Most items are priced in the lower to mid-range, but the long tail indicates the presence of high-priced products

---

**Total Sales Value:**  
Sales value shows a broad spread with a solid skew to the right. 

---

**Discount Rate (%):**  
Unlike the other columns, the discount rate, ranging from around 5% to nearly 20%, and appears slightly left-skewed. This indicates that most discounts hover near the upper end of the range. More discounts are given when product size is high

---

**Discount Amount:**  
The discount amount follows a right-skewed distribution, with many transactions having modest discount values and fewer instances of very high discounts.

---

**Net Sales Value:**  
Net sales value retains the right-skewed nature of total sales, as it’s directly impacted by both the original sales amount and the discount applied. Most transactions fall in the lower value range.

### Standardisation of Numerical Values

```python
# List of numerical columns
num_cols = [
    "Volume", "Avg Price", "Total Sales Value", 
    "Discount Rate (%)", "Discount Amount", "Net Sales Value"
]

zscore_df = data[num_cols].apply(lambda x: (x - x.mean()) / x.std())

zscore_df.columns = [f"{col}_zscore" for col in zscore_df.columns]

df_with_zscores = pd.concat([data[num_cols], zscore_df], axis=1)

df_with_zscores.head()
```

```python

plt.figure(figsize=(18, 16))

for i, col in enumerate(zscore_df.columns, 1):
    plt.subplot(3, 2, i)
    sns.histplot(zscore_df[col], bins=30, kde=True, color='coral')
    plt.title(f'Histogram of {col}')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
