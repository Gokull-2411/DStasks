```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data=pd.read_csv('Online retail.csv')
```

```python
data.head()
```

```python
data.columns = ['Items'] + [f'Unused_{i}' for i in range(1, len(data.columns))]
data= data[['Items']]
row=pd.DataFrame(
    [
        {
            'Items':'shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'
        }
    ]
)
data=pd.concat([data,row],ignore_index=True)
data
```

```python
data['Items']=data['Items'].astype(str).apply(lambda x: x.strip('"').split(','))
data
```

```python
enc=TransactionEncoder()
enc_arr=enc.fit_transform(data['Items'])
```

```python
enc.columns_
```

```python
data_enc=pd.DataFrame(enc_arr,columns=enc.columns_)
data_enc
```

```python
frequent_itemsets=apriori(data_enc,min_support=0.01,use_colnames=True)
frequent_itemsets
```

```python
assoc_rules=association_rules(frequent_itemsets,min_threshold=0.3)
assoc_rules
```

```python
rules=assoc_rules[assoc_rules['lift']>1]
```

```python
rules
```

```python
rules.sort_values(by='confidence', ascending=False).head(10)
```

### 1. What is Lift and Why Is It Important in Association Rules?

**Lift** measures how much more likely two items are to be purchased together compared to being bought independently. It is calculated as the ratio of the observed co-occurrence of items to the expected co-occurrence if they were independent. A lift value greater than 1 indicates a positive association, while less than 1 implies a negative association. Lift is important because it helps identify truly meaningful rules by filtering out coincidental co-purchases. Unlike confidence, lift adjusts for item popularity, making it a better metric for uncovering strong and non-trivial relationships between products in market basket analysis.


### 2. What is Support and Confidence? How Are They Calculated?

**Support** measures how frequently an itemset appears in the dataset. It is calculated as the proportion of transactions that contain the itemset:
Support(A,B) = (Transactions containing A and B) / (Total transactions)
**Confidence** measures the likelihood that a rule’s consequent appears in transactions that contain the antecedent:
Confidence(A ⇒ B) = Support(A and B) / Support(A)
Support helps identify common itemsets, while confidence reflects the strength of occurance. Both are essential in evaluating rule relevance association rule mining, guiding which rules to keep.


### 3. What Are Some Limitations or Challenges of Association Rule Mining?

Association rule mining has several limitations such as, 
It often generates a large number of rules, many of which may be trivial or uninteresting, leading to analysis overload. 

Setting appropriate thresholds for support and confidence is challenging—too low may yield noise, while too high may miss useful patterns. 

It also struggles with sparse or high-dimensional data, where meaningful patterns are rare.  

Popular items may dominate results unless lift or other measures are used, making it harder to detect less obvious but valuable associations.
