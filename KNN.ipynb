### Split the data

```python
X=data.iloc[:,1:17]
X
```

```python
Y=data.iloc[:,17:18]
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score,classification_report
data['type'].nunique()
```

### Elbow Method

```python
error_rate=[]
k=range(1,10)
for i in k:
    kn=KNeighborsClassifier(n_neighbors=i)
    kn.fit(xtrain,ytrain.squeeze())
    pred=kn.predict(xtest)
    error=1-accuracy_score(ytest,pred)
    error_rate.append(error)
    
error_rate
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k, error_rate, marker='o', linestyle='--', color='blue')
plt.title('Elbow Method for Optimal k in KNN')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error Rate')
plt.xticks(k)
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=5,metric='euclidean')
knn.fit(xtrain,ytrain)
```

```python
yhat_train=knn.predict(xtrain)
```

```python
print(classification_report(ytrain,yhat_train,zero_division=1))
```

```python
yhat_test=knn.predict(xtest)
print(classification_report(ytest,yhat_test,zero_division=1))
```

Since the model is trained on a very small dataset, the accuracy of the model in training and testing is high.

**1. Key Hyperparameters in KNN (K-Nearest Neighbors):**
The main hyperparameters in KNN include k (number of neighbors), which determines how many nearby points influence the prediction, distance metric (e.g., Euclidean), and weight function (uniform or distance-based). These parameters significantly affect model performance, influencing overfitting/underfitting, classification accuracy, and sensitivity to noisy data or class imbalances.

**2. Distance Metrics in KNN:**
KNN can use various distance metrics to measure similarity between data points. Common ones include Euclidean distance(straight-line distance), Manhattan distance (sum of absolute differences), Minkowski distance (a generalization) The choice of metric impacts how neighbors are selected and affects prediction accuracy.
