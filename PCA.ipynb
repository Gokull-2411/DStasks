### Data Standardisation

```python
data_pca=pd.read_csv("wine.csv")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
data_pca_scaled=scaler.fit_transform(data_pca)
data_pca_scaled
```

### Find PCs

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=14)
pcomps=pca.fit_transform(data_pca_scaled)
pcomps
```

```python
ratio=pca.explained_variance_ratio_
ratio
```

```python
var=np.cumsum(ratio)*100
var
```

### Components with respect to Variance

```python
plt.plot(var,linestyle='dotted',marker='H')
plt.ylabel("Information retained")
plt.xlabel("No of PCs")
```

```python
pcomps
```

```python
from sklearn.manifold import TSNE
```

```python
tsne =TSNE(n_components=2,perplexity=5)
tcomps=tsne.fit_transform(data_pca_scaled)
```

```python
plt.scatter(tcomps[:,0], tcomps[:,1],c=data['Type'])
plt.show()
```

```python
data_pca_final=pd.DataFrame(
    {
        'PC1':pcomps[:,0],
        'PC2':pcomps[:,1]
    }
)
data_pca_final
```

### KMeans with Original data 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss = []
for k in range(1, 13):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1, 13), wcss, marker='o')
plt.title("KMeans Elbow Curve")
plt.xlabel("K")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(data)
data['Kmean_label']=kmeans_labels
```

```python
kmeans_silhouette = silhouette_score(data, kmeans_labels)
print(f"KMeans Silhouette Score: {kmeans_silhouette:.3f}")
```

### 
With K means with original data, we get a silhouette score pf 0.56 and the wcss is more inclined towards 4 number of clusters in the Elbow Method

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, perplexity=5)
tcomps = tsne.fit_transform(data)
tcomps
```

```python
plt.figure(figsize=(6,3))

plt.scatter(tcomps[:,0], tcomps[:,1], 
            cmap='viridis', 
            c=data['Type'],
            alpha=0.5,
            edgecolors='black'
           )
plt.show()
```

### Kmeans with PCA Data

```python
wcss = []
for k in range(1, 13):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_pca_final)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1, 13), wcss, marker='o')
plt.title("KMeans Elbow Curve")
plt.xlabel("K")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(data_pca_final)
data['Kmean_label']=kmeans_labels
```

```python
kmeans_silhouette = silhouette_score(data_pca_final, kmeans_labels)
print(f"KMeans Silhouette Score: {kmeans_silhouette:.2f}")
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, perplexity=5)
tcomps = tsne.fit_transform(data_pca_final)
tcomps
```

```python
plt.figure(figsize=(6,3))

plt.scatter(tcomps[:,0], tcomps[:,1], 
            cmap='viridis', 
            c=data['Type'],
            alpha=0.5,
            edgecolors='black'
           )
plt.show()
```

Silhouette Score has imporved well enough and the data is well clusters into 3 clusters almost 100%. 

### Comparison and Conclusion

When comparing the clustering results obtained from the original dataset and the PCA-transformed data, we find that applying clustering after dimensionality reduction provides a more structured outcome. In the original dataset, clustering using KMeans can suffer due to high dimensionality, noise, or correlated features. On the other hand, clustering on PCA-transformed data simplifies the feature space, making it easier for the algorithm to identify patterns and groupings. This was evident in the notebook where clustering was performed after PCA, and the silhouette score was computed to evaluate the results.

One key observation is that while both approaches used KMeans for clustering, the PCA-transformed data resulted in better-defined clusters. The clusters obtained after PCA were more compact and well-separated, which was further validated through visualization techniques like t-SNE. This difference can be understood to PCA’s ability to capture the most important variance in the data while eliminating redundant features. Such transformation often helps unsupervised algorithms like KMeans to perform better.

### Practical Implication of PCA

PCA followed by clustering is useful when dealing with complex, multi-feature datasets. It helps uncover hidden structures while keeping computational cost low. On the other hand, if the dataset is already concise and the interpretability of features is a priority, clustering on the original data might be more suitable.

It is recommended to apply PCA before clustering when the dataset is high-dimensional, noisy, or exhibits multicollinearity. However, if the goal is to maintain direct interpretability or when working with already clean and small datasets, clustering directly on the original features is recommended
